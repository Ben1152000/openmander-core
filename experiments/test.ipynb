```python
%pip install openmander
import openmander as om
from pathlib import Path
from IPython.display import SVG, display

# Directory containing this script
OPENMANDER_ROOT = Path.cwd().parent

# Build paths relative to the root
BASE_PATH = OPENMANDER_ROOT / "experiments" / "packs"
SVG_PATH  = OPENMANDER_ROOT / "experiments" / "images"
```

```python
state = 'NY'
num_districts = 6

pack_path = om.build_pack(state, path=str(BASE_PATH), verbose=1)
# pack_path = BASE_PATH / (state + '_2020_pack')

plan = om.Plan(om.Map(str(pack_path)), num_districts)

plan.randomize()
plan.equalize(series="T_20_CENS_Total", tolerance=0.0001, max_iter=1000)

plan.to_csv(str(SVG_PATH / (state + ".svg")), )

plan.anneal_balance(
    series="T_20_CENS_Total",
    max_iter=100_000 * num_districts,
    initial_temp=100,
    final_temp=0.01,
    boundary_factor=0.01
)

plan.to_svg(str(SVG_PATH / (state + ".svg")), color_partisan=True)
display(SVG(filename=SVG_PATH / (state + ".svg")))
```
