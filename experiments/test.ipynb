```python
%pip install openmander
import openmander as om
from IPython.display import SVG, display

base_path = '/home/bzd2/openmander-core/experiments/packs/'
svg_path = "/home/bzd2/openmander-core/experiments/images/"

def display_plan(plan: om.Plan, name):
    plan.to_svg(svg_path + name)
    display(SVG(filename=svg_path + name))
```

```python
state = 'NY'
num_districts = 26

# pack_path = om.build_pack(state, path=base_path, verbose=1)
pack_path = base_path + state + '_2020_pack'

plan = om.Plan(om.Map(pack_path), num_districts)

plan.randomize()
plan.equalize(series="T_20_CENS_Total", tolerance=0.0001, max_iter=1000)
plan.anneal_balance(
    series="T_20_CENS_Total",
    max_iter=2_000_000 * num_districts,
    initial_temp=100,
    final_temp=0.01,
    boundary_factor=0.01
)

plan.to_svg(svg_path + state + ".svg", color_partisan=True)
display_plan(plan, state)
```
