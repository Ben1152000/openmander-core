```python
%pip install openmander
import openmander as om
from IPython.display import SVG, display

base_path = '/Users/benjamin/Programming/projects/openmander/packs/'
svg_path = base_path + 'plan.svg'

def display_plan(plan: om.Plan):
    plan.to_svg(svg_path)
    display(SVG(filename=svg_path))

# pack_path = om.build_pack('CT', path=base_path, verbose=1)
pack_path = '/Users/benjamin/Programming/projects/openmander/packs/CT_2020_pack'
map = om.Map(pack_path)
```

```python
plan = om.Plan(map, num_districts=5)
plan.series()

map.to_svg(svg_path, layer="tract", series="outer_perimeter_m")
display(SVG(filename=svg_path))
```

```python
plan = om.Plan(map, num_districts=5)
plan.randomize()
print(plan.compute_metric(om.Metric.compactness_polsby_popper()))
print(plan.compute_metric(om.Metric.compactness_schwartzberg()))
print(plan.compute_metric(om.Metric.competitiveness_binary(dem_series="E_20_PRES_Dem", rep_series="E_20_PRES_Rep", threshold=0.05)))
print(plan.compute_metric(om.Metric.competitiveness_quadratic(dem_series="E_20_PRES_Dem", rep_series="E_20_PRES_Rep", threshold=0)))
print(plan.compute_metric(om.Metric.competitiveness_gaussian(dem_series="E_20_PRES_Dem", rep_series="E_20_PRES_Rep", sigma=0.04)))
print(plan.compute_metric(om.Metric.population_deviation(pop_series="T_20_CENS_Total")))
print(plan.compute_metric(om.Metric.population_deviation_smooth(pop_series="T_20_CENS_Total")))


display_plan(plan)
```

```python
plan.equalize(series="T_20_CENS_Total", tolerance=0.0001, max_iter=1000)
print(plan.compute_metric(om.Metric.compactness_polsby_popper()))

display_plan(plan)
```

```python
plan.anneal_balance(series="T_20_CENS_Total", max_iter=100_000_000, initial_temp=100, final_temp=0.01, boundary_factor=0.01)
print(plan.compute_metric(om.Metric.compactness_polsby_popper()))

display_plan(plan)

# metric = om.Metric.population_deviation(series="T_20_CENS_Total")
# print(plan.compute_metric(metric))


# print(plan.compute_metric(om.Metric.population_deviation("T_20_CENS_Total")))
```
